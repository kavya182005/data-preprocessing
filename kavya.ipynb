```python







from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Initialize the LabelEncoder
encoder = LabelEncoder()

# Verify column names
print(df.columns)

# Apply Label Encoding only if columns exist
if 'Size' in df.columns:
    df['Size'] = encoder.fit_transform(df['Size'])
else:
    print("Column 'Size' does not exist in the DataFrame.")

if 'Frequency of Purchase' in df.columns:
    df['Frequency of Purchase'] = encoder.fit_transform(df['Frequency of Purchase'])
else:
    print("Column 'Frequency of Purchase' does not exist in the DataFrame.")

if 'Subscription Status' in df.columns:
    df['Subscription Status'] = encoder.fit_transform(df['Subscription Status'])
else:
    print("Column 'Subscription Status' does not exist in the DataFrame.")

# Apply One-Hot Encoding for categorical columns
categorical_columns = [
    'Gender', 'Item Purchased', 'Category', 'Location', 
    'Color', 'Season', 'Payment Method', 'Discount Applied', 
    'Promo Code Used', 'Shipping Type', 'Preferred Payment Method'
]

for col in categorical_columns:
    if col in df.columns:
        df = pd.get_dummies(df, columns=[col], drop_first=True)
    else:
        print(f"Column '{col}' does not exist in the DataFrame.")

# Verify the changes
print(df.dtypes)  # Check data types
print(df.head())  # Preview the first few rows

```

```python

# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Load the dataset
file_path = 'shopping.csv'  # Ensure this matches your file name
try:
    df = pd.read_csv(file_path)
    print("File loaded successfully!")
except FileNotFoundError:
    print(f"File not found: {file_path}. Please check the file name and path.")
    raise

# Display the first few rows to verify data
print("Dataset preview:")
print(df.head())

# ---- Feature Engineering---

# 1. Normalize numerical columns
numerical_columns = ['Age', 'Purchase Amount (USD)', 'Previous Purchases']
if all(col in df.columns for col in numerical_columns):
    scaler = StandardScaler()
    df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
    print("Numerical columns normalized.")
else:
    print("Some numerical columns are missing. Check your dataset.")

# 2. Handle categorical columns with OneHotEncoding
categorical_columns = ['Gender', 'Category', 'Location', 'Size', 'Color', 'Season', 
                       'Subscription', 'Status', 'Payment Method', 'Shipping Type', 
                       'Preferred Payment Method']
missing_categorical = [col for col in categorical_columns if col not in df.columns]
if missing_categorical:
    print(f"Missing categorical columns: {missing_categorical}")
else:
    ohe = OneHotEncoder(sparse=False, drop='first')
    encoded_categorical = pd.DataFrame(ohe.fit_transform(df[categorical_columns]), 
                                        columns=ohe.get_feature_names_out(categorical_columns))
    df = pd.concat([df.drop(columns=categorical_columns), encoded_categorical], axis=1)
    print("Categorical columns encoded.")

# 3. Encode binary columns
binary_columns = ['Discount Applied', 'Promocode Use']
for col in binary_columns:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: 1 if x == 'Yes' else 0)
        print(f"Binary column '{col}' encoded.")
    else:
        print(f"Binary column '{col}' is missing.")

# 4. Convert 'Frequency of Purchases' into ordinal categories
if 'Frequency of Purchases' in df.columns:
    frequency_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
    df['Frequency of Purchases'] = df['Frequency of Purchases'].map(frequency_mapping)
    print("'Frequency of Purchases' converted to ordinal.")
else:
    print("'Frequency of Purchases' column is missing.")

# 5. Convert 'Review Rating' to numeric
if 'Review Rating' in df.columns:
    df['Review Rating'] = pd.to_numeric(df['Review Rating'], errors='coerce')
    print("'Review Rating' converted to numeric.")
else:
    print("'Review Rating' column is missing.")

# 6. Drop unnecessary columns (e.g., Customer ID)
if 'Customer ID' in df.columns:
    df = df.drop(columns=['Customer ID'])
    print("'Customer ID' column dropped.")

# Save the processed dataset
processed_file_path = 'processed_shopping.csv'
df.to_csv(processed_file_path, index=False)
print(f"Feature engineering completed. Processed file saved at: {processed_file_path}")

```
